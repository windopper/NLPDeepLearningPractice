가중치 출력


```python
import tensorflow as tf

mnist = tf.keras.datasets.mnist
(x_train, y_train), (x_test, y_test) = mnist.load_data()

x_train, x_test = x_train/255.0, x_test/255.0

model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(512, activation=tf.nn.relu),
    tf.keras.layers.Dense(10, activation=tf.nn.softmax)
])

weights = model.get_weights()

print(weights)
print(len(weights))
```

첫번째 어레이 
784x512개의 값을 갖는 2차원 어레이, 입력층과 은닉층을 연결하는 가중치

두번째 어레이
512개의 0으로 이루어져 있으면 은닉층의 편향값을 나타냄

세번째 어레이 512x10개의 값을 같는 2차원 어레이 , 은닉층과 출력층을 연결하는 가중치

네번째 어레이 10개의 0으로 이루어져 있으며 출력층의 편향 값을 나타냄

```python
print(weights[0].shape)
print(weights[1].shape)
print(weights[2].shape)
print(weights[3].shape)
```

웨이트 저장하기

numpy 의 savetxt 함수를 사용하면 저장가능


```python
import numpy as np

np.savetxt('weights[0].csv', weights[0])
np.savetxt('weights[1].csv', weights[1])
np.savetxt('weights[2].csv', weights[2])
np.savetxt('weights[3].csv', weights[3])
```
